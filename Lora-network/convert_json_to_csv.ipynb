```python
# import required module
import os
import re
import pandas as pd

if not os.path.exists("/content/data_csv"):
  os.mkdir("/content/data_csv")
# assign directory
directory = '/content'

# iterate over files in
# that directory
for filename in os.listdir(directory):
	f = os.path.join(filename)
	# checking if it is a file
	if os.path.isfile(f):
		print(f)
```

```python


for filename in os.listdir(directory):
  f = os.path.join(filename)
  if os.path.isfile(f):
    ## creation des dico vide
    dico={} 
    dico_seq={} 

    ##ouverture du fichier fasta avant utilisation de la boucle pour le remplissage des dico 
    F=open(f, 'r') 

    ## j est un compteur pour le dico contenant l'ensemble des sequences  
    j=0 
    for i in F:
        clean = i.strip() ## eleminer les /n situe à la fin de la sequence 
        i = re.sub(r"\s+", " ", i)        

        if "received_at" in i :
          x=i
          x=x.replace('"received_at": "','')
          x=x.replace('",','')
          dico={}
          dico['date']=x
        if "do1" in i : 
          x=i
          x=x.replace('"do1":','')
          x=x.replace(',','')
          dico['do1']=x
        if "do2" in i : 
          x=i
          x=x.replace('"do2":','')
          x=x.replace(',','')
          dico['do2']=x
        
        if "ec" in i : 
          x=i
          x=x.replace('"ec":','')
          x=x.replace(',','')
          dico['ec']=x


        if "ph" in i : 
          x=i
          x=x.replace('"ph":','')
          x=x.replace(',','')
          dico['ph']=x

        if 'temp"' in i : 
          x=i
          x=x.replace('"temp":','')
          x=x.replace(',','')
          dico['temp']=x

        if "tempbox" in i : 
          x=i
          x=x.replace('"tempbox":','')
          x=x.replace(',','')
          dico['tempbox']=x

        if "turb" in i : 
          x=i
          x=x.replace('"turb":','')
          dico['turb']=x
          #flag=False
          dico_seq[j]=dico
          j=j+1
        
    F.close()

    data = pd.DataFrame(dico_seq)
    df=data.T

    df.to_csv("/content/data_csv/"+f+'.csv', sep='\t')
```
